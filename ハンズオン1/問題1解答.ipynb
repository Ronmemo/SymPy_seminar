# 方程式

(1)
$$
x^2-12x+8 = 0
$$

```python
eq = sym.Eq(x**2-12*x+8)
eq
```

```python
sym.solve(eq, x)
```

(2)
$$
x^3 + \frac{1}{x^3} = 52
$$
http://www.k-kyogoku.com/cn81/cn108/pg1758.html

```python
eq = sym.Eq(x**3 - 1/x**3, 52)
eq
```

```python
sym.solve(eq, x)
```

(3)
$$
a x^3 + bx^2 + cx + d = 0
$$

```python
# 3次方程式
eq = sym.Eq(a * x ** 3 + b * x ** 2 + c * x + d)
eq
```

```python
# 方程式を解く
sym.solve(eq, x)
```

(4)
$$
x + y = 5 \\
x^3 + y^3 = 35
$$
http://kamelink.com/blog/tag/連立方程式/

```python
eq1 = sym.Eq(x + y, 5)
eq1
```

```python
eq2 = sym.Eq(x**3 + y**3, 35)
eq2
```

```python
sym.solve([eq1, eq2])
```

(5)
$$
x+y=12 \\
y+z=13 \\
z+x=15
$$

```python
eq1 = sym.Eq(x + y, 12)
eq1
```

```python
eq2 = sym.Eq(y + z, 13)
eq2
```

```python
eq3 = sym.Eq(z + x, 15)
eq3
```

```python
sym.solve([eq1, eq2, eq3])
```

# 微分

(1)
$$
\frac{d}{dx}(x^2-12x+8)
$$

```python
F = x**2 - 12*x + 8
F
```

```python
plot(F, (x, -10, 10))
```

```python
diff_F = sym.diff(F)
diff_F
```

```python
mx = sym.solve(sym.Eq(diff_F, 0))
mx
```

```python
my = F.subs(x, mx[0])
my
```
